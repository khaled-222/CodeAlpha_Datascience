```python
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
x = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[x]))
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum().sum()
```

```python
df.describe()
```

```python
df["Frequency"].unique()
```

```python
df["Area"].unique()
```

```python
df["Region"].nunique()
```

```python
df.drop(columns=["Frequency"],axis=1,inplace=True)
df.head()
```

```python
df["Date"] = df["Date"].str.strip()
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Keep only Year + Month
df["Date"] = df["Date"].dt.strftime("%Y")
display(df.head())
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Date"], df[" Estimated Unemployment Rate (%)"], color='skyblue')
plt.title("Unemployment Percentage per Year")
plt.xlabel("Year")
plt.ylabel("Unemployment Percentage (%)")
plt.show()
```

```python
Rural_df = df[df["Area"] == "Rural"]
Urban_df = df[df["Area"] == "Urban"]
```

```python
plt.figure(figsize=(8,5))
plt.bar(Rural_df["Date"], Rural_df[" Estimated Unemployment Rate (%)"], color='blue')
plt.title("Rual Unemployment")
plt.xlabel("Year")
plt.ylabel("Unemployment (%)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(Urban_df["Date"], Urban_df[" Estimated Unemployment Rate (%)"], color='green')
plt.title("city Unemployment")
plt.xlabel("Year")
plt.ylabel("Unemployment (%)")
plt.grid(True)
plt.show()
```
